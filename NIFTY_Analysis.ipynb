```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Download NIFTY 50 data
nifty = yf.download("^NSEI", start="2023-01-01", end="2024-01-01")

nifty.head()
```

```python
# Remove missing values
nifty = nifty.dropna()

# Calculate daily returns
nifty["Daily Return"] = nifty["Close"].pct_change()

nifty.tail()
```

```python
plt.figure(figsize=(10,4))
plt.plot(nifty.index, nifty["Close"])
plt.title("NIFTY 50 Closing Price")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

```python
avg_return = nifty["Daily Return"].mean()
volatility = nifty["Daily Return"].std()

avg_return, volatility
```
